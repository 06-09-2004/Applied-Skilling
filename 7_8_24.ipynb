```python
import matplotlib.pyplot as plt
#defining categories and values for two groups
categories=['Transport','income tax','Housing']
values1=[15,24,30]
values2=[20,18,25]
#creating the first set of bars (group1) without any offset
plt.bar(categories,values1,label='first half',color='blue')
#creating the second set of bars (group2) plotted eith offset
#this makes group 2 bars stacked on top of group 1 bars
plt.bar(categories,values2,bottom=values1,label='second half',color='pink')
#Adding labels to the axes
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Stacked Bar Graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[70,50,100,70,50,100,50,50,50,70,50]
plt.hist(x,orientation="vertical")
plt.title("Mark Analysis")
plt.ylabel("No of Students")
plt.xlabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#generate random exam scores(out of 100)
exam_scores=np.random.randint(0,100,150)
#Create a cumulative histogram
plt.hist(exam_scores,bins=20,cumulative=True,color='blue',edgecolor="black")
plt.xlabel('Exam Scores')
plt.ylabel('Cumulative Number of Students')
plt.title('Cumulative Histogram of Exam Scores')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#Sample data for two datasets
mark1=np.array([2,4,5,7,9,10,11,13,14,15])
mark2=np.array([6,7,8,10,11,12,13,14,15,16])
#creating a stacked histogram with different colors
plt.hist([mark1,mark2],bins=10,stacked=True,color=['pink','blue'],edgecolor="black")
plt.xlabel('Marks')
plt.ylabel('Count')
plt.title('stacked Histogram with Multiple Datasets')
plt.legend(['Mark Dataset 1', 'Mark Dataset 2'])
plt.show()
```
