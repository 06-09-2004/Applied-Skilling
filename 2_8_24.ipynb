```python
from matplotlib import pyplot as plt
import numpy as np
import math #needed for definition of pi
x=np.arange(0,math.pi*2,0.05)
y=2*(np.sin(x))
#print(x)
#print(y)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
import matplotlib.pyplot as plt
#example data
year=[1,2,3]
Veg_Rate=[2,4,6]
Milk_Rate=[1,3,5]
Fruit_Rate=[3,6,9]
#Plotting the data
line1, =plt.plot(year, Veg_Rate)
line2, =plt.plot(year, Milk_Rate)
line3, =plt.plot(year, Fruit_Rate)
#calling legend with explicitly listed artists and labels
plt.legend([line1, line2,line3],['vegetable','milk','Fruits'])
plt.xlabel('years')
plt.ylabel('rate')
#show the plot
plt.show()
print('Successfully placed a legend on the Axes...')
```

```python
import matplotlib.pyplot as plt
categories =['lion','horse','kangaroo']
values=[15,24,30]
plt.bar(categories,values,color='skyblue')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Basic Vertical Bar Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#defining categoties and their corresponding values for
categories=['lion','horse','kangaroo']
values1=[15,24,30]
values2=[20,18,25]
#setting the width of the bars
bar_width=0.35
#calculating bar positions for both groups
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1 + bar_width
#creating the first set of bars(group 1)
plt.bar(bar_positions1,values1,width=bar_width,label='2021',color='skyblue')
#create the second set of bars(group 2)next to the first set
plt.bar(bar_positions2,values2,width=bar_width,label='2022',color='yellow')
#Adding labels to the axes
plt.xlabel('categories')
plt.ylabel('values')
#adding a title to the group
plt.title('grouped bar graph')
#displaying a legend to identify the groups
plt.legend()
#showing the plot
plt.show()








```
